```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage import data
from skimage.feature import greycomatrix, greycoprops
from skimage import io, color, feature
```

```python
img = cv2.imread('/content/kucing.jpeg')
image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(image, cmap='gray')
plt.show()
```

```python
image
```

```python
distances = [1, 2, 3]
angles = [0, np.pi/4, np.pi/2, 3*np.pi/4]
```

```python
glcm = feature.greycomatrix(image, distances=distances, angles=angles, levels=256, symmetric=True, normed=True)

contrast = feature.greycoprops (glcm, 'contrast')
homogeneity = feature.greycoprops(glcm, 'homogeneity')
energy = feature.greycoprops(glcm, 'energy')
correlation = feature.greycoprops(glcm, 'correlation')
```

```python
print('Contrast: ', contrast)
print('Homogeneity: ', homogeneity)
print('Energy: ', energy)
print('Correlation: ', correlation)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_jarak = {'jarak_A (M)':['15', '14', '19', '18'], 'jarak_B (M)':['20', '25', '21', '23']}

df = pd.DataFrame(data_jarak)

df
```

```python
sns.lineplot(x="jarak_A (M)", y="jarak_B (M)", data=data_jarak)

plt.title('JARAK A VS JARAK B\n', size=16)
plt.ylabel('\n jarak B', size=14)
plt.xlabel('\n jarak A', size=14)
plt.show
```

```python
def euclidean_distance_timeseries(jarak_a, jarak_b):
  """
  Fungsi untuk menghitung jarak euclidean antara dua seri waktu.

  parameters:
  jarak_a (numpy.ndarray): Seri waktu pertama.
  jarak_b (numpy.ndarray): Seri waktu kedua.

  returns:
  float Jarak Euclidean antara dua seri waktu
  """

  # cek apakah kedua seri waktu memiliki panjang yang sama
  if len(jarak_a) != len(jarak_b):
    raise ValueError("kedua seri waktu memiliki panjang yang sama. ")

  # Hitung jarak Euclidean antara kedua seri waktu
  dist = np.sqrt(np.sum(np.power(jarak_a - jarak_b, 2)))

  return dist
```

```python
jarak_a = np.array([15, 14, 19, 18])
jarak_b = np.array([20, 25, 21, 23])
dist = euclidean_distance_timeseries(jarak_a, jarak_b)
print('akurasi jarak euclidean time series : ', dist) # output 13.228756555322953
```
